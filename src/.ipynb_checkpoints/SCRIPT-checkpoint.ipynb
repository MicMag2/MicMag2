# Define helperfunctions  (evtl auslagern oder verstecken)

```python
def plot_j(j, nx, ny):   
    
    lw = np.sqrt(j[:,:,0,0]**2+j[:,:,0,1]**2)
    
    x_ = np.linspace(0,nx, nx)
    y_ =np.linspace(0,ny, ny)
    x,y = np.meshgrid(y_, x_)
    
    c1 = plt.Circle((0.5*ny, 0.9*nx), r_contact/dx, alpha=0.5, color="green")
    c2 = plt.Circle((0.5*ny, 0.1*nx), r_contact/dx, alpha=0.25, color="red")
    c3 = plt.Circle((0.5*ny,0.5*nx, ), r_hole/dx, alpha=0.75,fill=False, linestyle="--", linewidth=2.5,color="black")
    #c3 = plt.Circle((75, 124), r_hole/dx, alpha=0.75,fill=False, linestyle="--", linewidth=2.5,color="black")
    
    fig2, ax = plt.subplots(figsize=(2,5))
    
    strm = plt.streamplot(x, y, j[:,:,0,1],j[:,:,0,0], linewidth = lw*0.35e-13, color =lw*0.3e-13,  cmap="autumn")
    plt.title("$j_\mathrm{IP}$ [a.u.]", fontsize=16)
    #cb = fig2.colorbar(strm.lines)
    #cb.ax.set_ylabel(r'$j_\mathrm{IP} [UNIT MISSING]$',fontsize=16)
    
    plt.yticks(ticks=[], labels=[])
    plt.xticks(ticks=[ ], labels=[])
    plt.ylabel("x", fontsize=16)
    plt.xlabel("y", fontsize=16)
    
    
    ax.add_patch(c1)
    ax.add_patch(c2)
    ax.add_patch(c3)
    plt.show()
    
def plot_m(mag, nx, ny):
    fig, ax = plt.subplots()
    ax2 = plt.imshow(mag[:,:,0,2], vmin=-1, vmax=1, cmap="RdYlBu")
    c1 = plt.Circle((0.5*ny, 0.9*nx), r_contact/dx, alpha=0.5, color="green")
    c2 = plt.Circle((0.5*ny, 0.1*nx), r_contact/dx, alpha=0.25, color="red")
    c3 = plt.Circle((0.5*ny,0.5*nx, ), r_hole/dx, alpha=0.75,fill=False, linestyle="--", linewidth=2.5,color="black")
    ax.add_patch(c1)
    ax.add_patch(c2)
    ax.add_patch(c3)
    
    cb = plt.colorbar(ax2)
    cb.ax.set_ylabel(r'$m_\mathrm{z} $',fontsize=16)
    plt.yticks(ticks=[], labels=[])
    plt.xticks(ticks=[ ], labels=[])
    plt.ylabel("x", fontsize=16)
    plt.xlabel("y", fontsize=16)
    plt.title(r"$m_\mathrm{z}$", fontsize=16)
    plt.show()

def plot_wire(nx, ny):
    plt.imshow
    fig, ax = plt.subplots()
    
    ax2 = plt.imshow(wire, vmin=-1, vmax=1, cmap="RdYlBu")
    c1 = plt.Circle((0.5*ny, 0.9*nx), r_contact/dx, alpha=0.5, color="green")
    c2 = plt.Circle((0.5*ny, 0.1*nx), r_contact/dx, alpha=0.25, color="red")
    c3 = plt.Circle((0.5*ny,0.5*nx, ), r_hole/dx, alpha=0.75,fill=False, linestyle="--", linewidth=2.5,color="black")
    ax.add_patch(c1)
    ax.add_patch(c2)
    ax.add_patch(c3)
    
    cb = plt.colorbar(ax2)
    cb.ax.set_ylabel(r'$m_\mathrm{z} $',fontsize=16)
    plt.yticks(ticks=[], labels=[])
    plt.xticks(ticks=[ ], labels=[])
    plt.ylabel("x", fontsize=16)
    plt.xlabel("y", fontsize=16)
    plt.title(r"$m_\mathrm{init}$", fontsize=16)
    plt.show()
    
def plot_mag(solver, xlist, ylist, qlist):
    
    solver.state.M.normalize(1)
    mag = solver.state.M.to_numpy()
    fig, ax = plt.subplots()
    ax2 = plt.imshow(mag[:,:,0,2], vmin=-1, vmax=1, alpha=1.0, cmap="RdYlBu")
    c1 = plt.Circle((0.5*solver.mesh.num_nodes[1] , 0.9*solver.mesh.num_nodes[0]), r_contact/dx, alpha=0.5, color="green")
    c2 = plt.Circle((0.5*solver.mesh.num_nodes[1], 0.1*solver.mesh.num_nodes[0]), r_contact/dx, alpha=0.25, color="red")
    c3 = plt.Circle((0.5*solver.mesh.num_nodes[1],0.5*solver.mesh.num_nodes[0], ), r_hole/dx, alpha=0.75,fill=False, linestyle="--", linewidth=2.5,color="black")
    ax.add_patch(c1)
    ax.add_patch(c2)
    ax.add_patch(c3)
    
    cb = plt.colorbar(ax2)
    cb.ax.set_ylabel(r'$m_\mathrm{z} $',fontsize=16)
    plt.yticks(ticks=[], labels=[])
    plt.xticks(ticks=[ ], labels=[])
    plt.ylabel("x", fontsize=16)
    plt.xlabel("y", fontsize=16)
    
    solver.state.M.normalize(Msat)
    alphas = np.linspace(0.25, 1.0, len(Q_list))
    if len(qlist)>1:
        plt.scatter(np.array(x_poslist[:-1])/dx, np.array(y_poslist[:-1])/dy, alpha=alphas, color="darkorange", s=np.array(qlist[:-1])*40)
    
    #plt.scatter(np.array(xlist)/dx, np.array(ylist)/dy, alpha=alphas, color="darkorange")#, s=np.array(qlist)*40)
    plt.title(r"$m_\mathrm{t}$ & $r_\mathrm{sk}$, t = "+str(round(solver.state.t*1e9, 1))+" ns", fontsize=16)
    
    #plt.show()
    
    solver.state.M.normalize(Msat)
    display.display(fig)
    display.clear_output(wait=True)
def track_skyrmion():
    q_dens = np.squeeze(solver.state.Q_density.to_numpy()[:,:,0])
    x_ = np.linspace(0, nx*dx, nx)
    y_ = np.linspace(0, ny*dy, ny) 
    x,y = np.meshgrid(y_, x_)
    x_sk = np.average(x, weights=q_dens)
    y_sk = np.average(y, weights=q_dens)

    return x_sk,y_sk
```

# Define problem
### We want to investigate a wire (length = 500 nm, width = 200 nm, thickness = 2nm), which is contacted from top with cylindrical electrodes, and has also a cylindrical hole in it

# Define 3D Mesh (wire + contacts)

```python
width = 200e-9
length = 500e-9
dx,dy = 2e-9,2e-9                                 # definition of cell size
dz = 2e-9
nx,ny,nz = int(round(length/dx)),int(round(width/dy)),1   # definition of mesh size

mesh3D = RectangularMesh((nx,ny,nz),(dx,dy,dz),'')   # definition of the mesh with periodic boundaries in "xy" direction
```

# Define material geometry, parameter  (dict ausklappen)

```python
Msat = 5e5                                     # define saturation magnetisation (A/m)
A = 1.75e-11                                   # define exchange constant (J/m)
D = 1.75e-3                                    # define DMI constant (mJ/m^2)
alpha= 0.75                                    # define Gilbert damping parameter
k_uni = 360000                                 # define uniaxial anisotropy constant (J/m^3)

#definition of the SOT variables

xi =  0.0                                      # Field-like-SOT
alpha_H = 0.0025                               # spin hall ratio, 0.35

# definition of the material
material = Material({'Ms':Msat,
                     'alpha':alpha,
                     'A':A, 
                     'axis1':(0,0,1),
                     'axis2':(0,1,0),
                     'k_uniaxial':k_uni,
                     'k_cubic':0,
                     "sigma":7.75e6, 
                     "amr":0.1, 
                     "MST_alpha_hall":alpha_H,
                     "MST_xi":xi}) # define Material,
```

# Define the shape of 3D geometry (wire + contacts)

```python
r_contact= 20e-9
r_hole= 20e-9

wire = Cuboid((0,0,0), (nx*dx, ny*dy, nz*dz))
hole =  Cylinder((length/2, width/2, 0),(length/2, width/2, nz*dz), r_hole)


contact1 = Cylinder((0.9*length, width/2, 0),(0.9*length, width/2, nz*dz),r_contact)
contact2 = Cylinder((0.1*length, width/2,0),(0.1*length, width/2, nz*dz),r_contact)

wire_hole = wire.intersect(hole.invert())
wire_hole = wire_hole.intersect(contact1.invert())
wire_hole = wire_hole.intersect(contact2.invert())


#contact1 = Cylinder((0.9*length, width/2, 0),(0.9*length, width/2, nz*dz),r_contact)
#contact2 = Cylinder((0.1*length, width/2,0),(0.1*length, width/2, nz*dz),r_contact)

world3D = World(mesh3D, Body("contact1", Material.Au(), contact1 ),
              Body("contact2",Material.Au(), contact2 ),  
              Body("wire_hole", material, wire_hole))
```

# Calculate Current Path

```python
solver = create_solver(world3D, [CurrentPath], finescale = False, log=False)  # creates the solver, which can evolve the system. We include all necessary interactions

solver.state.M = (0,0,Msat)

solver.U_contact = 1e-3                         # [Volts]

plt.imshow(solver.state.Ms.to_numpy()[:,:], vmin=0)
plt.colorbar()
plt.show()
```

```python
#writeOMF("J_init.omf", solver.state.j)

j_np = solver.state.j.to_numpy()
        
plot_j(j_np, nx,ny)
print(np.sum(j_np))
#solver.removeModule("CurrentPath")
```

# Define solver and initialize magnetization  (Kommentar : Topology in der Liste ist merkwürdig)

```python
solver = create_solver(world3D, [MacroSpinTorque, ExchangeField, DMIField, AnisotropyField, StrayField, Topology ],finescale = False, log=False)  # creates the solver, which can evolve the system. We include all necessary interactions

solver.state.j = VectorField(mesh3D)
solver.state.j.from_numpy(j_np)

solver.state.topo_method="berg-luescher"     # Berg Luescher method to calculate the topological winding number

solver.state.Dx.fill((0, D, 0))              # we fill the DMI tenosr manually (interfacial DMI)
solver.state.Dy.fill((-D, 0, 0))             # this is interfacial DMI
solver.state.Dz.fill((0, 0, 0))               # you can customize the D tensor
solver.state.M = (0,0,Msat)                   # set m to (0,0,1)
solver.state.M.normalize(Msat)    
```

# Minimize state


```python
m_filename = "M_init.omf"

if True: # not os.path.isfile(m_filename):
    #if not os.path.isfile(m_filename):
    for i in range(int(0.650*nx), int(0.80*nx)):
        for j in range(int(0.4*ny), int(0.6*ny)):
            solver.state.M.set(i,j,0, (0,0,-Msat))
            #print(mesh3D.getPosition((i,j,0)))
    solver.state.M.normalize(1)    
    mag = solver.state.M.to_numpy()
    plot_m(mag, nx, ny)
    solver.state.M.normalize(Msat)    
    # minimise energy/ relax state
    solver.minimize(max_dpns=5, )
    writeOMF(m_filename, solver.state.M)



solver.state.M = readOMF(m_filename)

solver.state.M.normalize(1)    
mag = solver.state.M.to_numpy()
plot_m(mag, nx, ny)
solver.state.M.normalize(Msat)    
```

# Drive skyrmion (Kommentar: Position/Form vom Loch verändern)

```python
Q_list = []
x_poslist = []
y_poslist = []

Q = round(np.abs(solver.state.Q),3)
Q_list.append(Q)
x_sk, y_sk = track_skyrmion()
x_poslist.append(x_sk)
y_poslist.append(y_sk)
plot_mag(solver, x_poslist, y_poslist,Q_list)
dr = 1e-9
while Q > 0.1:
    while dr>1e-11:
    
        solver.solve(condition.Time(solver.state.t+1e-10))         # time integration of the system
        Q = round(np.abs(solver.state.Q),3)
        Q_list.append(Q)
        x_sk, y_sk = track_skyrmion()
        x_poslist.append(x_sk)
        y_poslist.append(y_sk)
        dr = np.sqrt((x_poslist[-1]-x_poslist[-2])**2+(y_poslist[-1]-y_poslist[-2])**2)
        plot_mag(solver, x_poslist, y_poslist,Q_list)
```
